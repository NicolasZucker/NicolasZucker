```python
import random
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
q=1;
a0=1+q;
N_step=6000000;
T_List=[0]
L_List=[10]
random.seed(1023)
TList=0
L=L_List[0]
T=T_List[0]
T_spend=[0]


for k in range(N_step) :
  if k%(int(N_step/30))==0:
    L=100
  u1=random.uniform(0,1)
  u2=a0*random.uniform(0,1)
  T = T + np.log(1/u1)/a0;
  T_List.append(T)
  T_spend.append(np.log(1/u1)/a0)
  if u2 < 1:
    #L'evènement q s'est produit
    L = L + 1;
  else :
    if L>1:
      L = L - 1;
  L_List.append(L)

plt.plot(T_List,L_List)
```

```python
plt.hist(L_List, weights=T_spend,bins=35)

```
